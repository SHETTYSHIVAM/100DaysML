```python
upper_limit = df['Height'].quantile(0.995) # above 99 percentile all observations are outlier
upper_limit
```

```python
lower_limit = df['Height'].quantile(0.01)
lower_limit
```

### Outliers

```python
df[(df['Height'] >= upper_limit) & (df['Height'] <= lower_limit)]
```

## Trimming

```python
new_df = df[(df['Height'] <= upper_limit) & (df['Height'] >= lower_limit)]
new_df
```

## Outliers

```python
df[(df['Height'] >= upper_limit) | (df['Height'] <= lower_limit)]
```

```python
new_df['Height'].describe()
```

```python
df['Height'].describe()
```

```python
sns.distplot(new_df['Height'])
```

```python
sns.boxplot(new_df['Height'])
```

# Capping --> Winsorization

```python
df['Height'] = np.where(
    df['Height'] > upper_limit,
    upper_limit, # or upper_limit + 1
    np.where(
        df['Height'] < lower_limit, 
        lower_limit, # or lower_limit - 1
        df['Height']
    )
)
```

```python
df['Height'].describe()
```

```python
sns.distplot(df['Height'])
```

```python
sns.boxplot(df['Height'])
```
