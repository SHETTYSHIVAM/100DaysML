# One-Hot-Encoding using sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df.iloc[:,0:4], df.iloc[:,-1], test_size = 0.2, random_state=2
)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel', 'owner']])
```

```python
x_train_new = ohe.transform(x_test[['fuel', 'owner']])
x_train_new
```

```python
# Ensure the arrays have matching dimensions before concatenation.
x_train_brand_km = x_train[['brand', 'km_driven']].values[:1626]  # Adjust to match the size of x_train_new
x_train_new_adjusted = x_train_new  # Assuming x_train_new has 1626 rows

# Concatenate the arrays
result_array = np.hstack((x_train_brand_km, x_train_new_adjusted))
result_array
```
